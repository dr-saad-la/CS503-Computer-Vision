```python
# Create a black grayscale image (2D array)
img = np.zeros((100, 100), dtype=np.uint8)

# Display the image
plt.imshow(img, cmap='gray')
plt.axis('off')  # Hide axis ticks
plt.show()
```

```python
# Create a black grayscale image (2D array)
img = np.zeros((100, 100), dtype=np.uint8)

# Set the first 50 rows to white (pixel value 255)
img[:50, :50] = 255

# Display the image
plt.imshow(img, cmap='gray')
plt.axis('off')  # Hide axis ticks
plt.show()
```

```python
# Create a white grayscale image (2D array) directly
img = np.ones((100, 100), dtype=np.uint8) * 255

# Display the image
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
# plt.axis('off')
plt.xticks([])
plt.yticks([])
plt.show()
```

### 1. Create Red Image 

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a red square image
red_img = np.zeros((100, 100, 3), dtype=np.uint8)
red_img[:, :] = [255, 0, 0]        # R, G, B

print(red_img.shape)      # (100, 100, 3)
plt.imshow(red_img)
plt.axis('off')
plt.show()
```

```python
green_img = np.zeros((100, 100, 3), dtype=np.uint8)
green_img[:, :] = [0, 255, 0]        # R, G, B

print(green_img.shape)      # (100, 100, 3)
plt.imshow(green_img)
plt.axis('off')
plt.show()
```

```python
blue_img = np.zeros((100, 100, 3), dtype=np.uint8)
blue_img[:, :] = [0, 0, 255]        # R, G, B

print(blue_img.shape)      # (100, 100, 3)
plt.imshow(blue_img)
plt.axis('off')
plt.show()
```
